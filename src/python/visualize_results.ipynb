```python
import os
path = os.getcwd()
path = path.split('interactions', 1)[0] + 'interactions'
print(path)
```

```python
#loading results data cicero with information if there was a pair from chromatin loops data to intersect with 
import pandas as pd

cicero = pd.read_csv(path+'/results/whole_data/cicero_conns.csv')
cicero_long = pd.read_csv(path+'/results/whole_data/cicero_conns_long.csv')
#cicero10 = pd.read_csv(path+'/results/whole_data/cicero_conns_intersected_range10.csv')
chromatin_loops = pd.read_excel(path+'/data/muszka/TableS2_Dmel_loops.xlsx')
chromatin_loops_bigger_set = pd.read_csv(path+'/data/muszka/long_and_short_range_loops_D_mel_formatted.csv')
cicero.head(10)
```

```python
results = pd.read_csv(path+'/results/whole_data/intersected_cicero_loops.csv')
results_bigger = pd.read_csv(path+'/results/whole_data/intersection_results_bigger_set.csv')
results_long = pd.read_csv(path+'/results/whole_data/intersection_results_cicero_long.csv')
```

```python
results.head(10)
```

```python
import numpy as np
cicero['middle1'] = (cicero['end1'] + cicero['start1'])/2
cicero['middle2'] = (cicero['end2'] + cicero['start2'])/2
cicero['dist'] = abs(cicero['middle1'] - cicero['middle2'])
cicero['log_dist'] = np.log10(cicero['dist'])

cicero_long['middle1'] = (cicero_long['end1'] + cicero_long['start1'])/2
cicero_long['middle2'] = (cicero_long['end2'] + cicero_long['start2'])/2
cicero_long['dist'] = abs(cicero_long['middle1'] - cicero_long['middle2'])
cicero_long['log_dist'] = np.log10(cicero_long['dist'])

chromatin_loops_bigger_set['middle1'] = (chromatin_loops_bigger_set['end1'] + chromatin_loops_bigger_set['start1'])/2
chromatin_loops_bigger_set['middle2'] = (chromatin_loops_bigger_set['end2'] + chromatin_loops_bigger_set['start2'])/2
chromatin_loops_bigger_set['dist'] = abs(chromatin_loops_bigger_set['middle1'] - chromatin_loops_bigger_set['middle2'])

cicero.head(10)
```

```python
cicero_y = cicero.loc[results_bigger['cicero_index']]
cicero_n = cicero[~cicero.index.isin(results_bigger['cicero_index'])]
```

```python
seed = 42
chromatin_loops_bigger_set['found_by_cicero'] = [0]*chromatin_loops_bigger_set.shape[0]
chromatin_loops_bigger_set.loc[results_bigger['loop_index'], 'found_by_cicero'] = 1
chromatin_loops_bigger_set['score'] = np.random.normal(loc=0, scale=0.4, size=chromatin_loops_bigger_set.shape[0])
chromatin_loops_bigger_set.loc[results_bigger['loop_index'], 'score'] = results_bigger['score']
chromatin_loops_bigger_set.head()
```

```python

plt.figure(figsize=(20,10))

# Assuming the two unique values in 'found_by_cicero' are 'value1' and 'value2'
loops_in_results = chromatin_loops_bigger_set[chromatin_loops_bigger_set['found_by_cicero'] == 1]
loops_outside_results = chromatin_loops_bigger_set[chromatin_loops_bigger_set['found_by_cicero'] == 0]

plt.scatter(np.log10(loops_in_results['dist']), loops_in_results['score'], color='#ffa600', s=30, label='found by cicero')
plt.scatter(np.log10(loops_outside_results['dist']), loops_outside_results['score'], color='#003f5c', s=30, label='not found by cicero')

plt.xlabel('Log10 loop size')  # x-axis label
plt.ylabel('Cicero score (or random value)')  # y-axis label
plt.legend()  # to show the legend
```
