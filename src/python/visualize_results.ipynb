```python
import numpy as np
cicero['middle1'] = (cicero['end1'] + cicero['start1'])/2
cicero['middle2'] = (cicero['end2'] + cicero['start2'])/2
cicero['dist'] = abs(cicero['middle1'] - cicero['middle2'])
cicero['log_dist'] = np.log(cicero['dist'])
cicero.head(10)
```

```python
cicero_y = cicero[cicero['intersection'] == 'y']
cicero_n = cicero[cicero['intersection'] == 'n']
cicero10_y = cicero10[cicero10['intersection'] == 'y']
cicero10_n = cicero10[cicero10['intersection'] == 'n']
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plt.scatter(cicero_n['log_dist'], cicero_n['score'], s=1)
```
